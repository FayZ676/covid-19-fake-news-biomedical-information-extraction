```python
import pandas as pd
import en_ner_bionlp13cg_md
import en_ner_bc5cdr_md
import en_ner_jnlpba_md
import en_ner_craft_md
import spacy
```

```python
df = pd.read_csv('data.csv')
df = df.dropna(subset=['text', 'label'])
df.head()
```

```python
# Get Spacey SCI named entities
nlp5 = spacy.load('medical-ner') # Custom medical-ner model
nlp4 = en_ner_bionlp13cg_md.load()
nlp3 = en_ner_bc5cdr_md.load()
nlp2 = en_ner_jnlpba_md.load()
nlp1 = en_ner_craft_md.load()

# Entities in craft
df['GGP'] = 0
df['SO'] = 0
df['TAXON'] = 0
df['CHEBI'] = 0
df['GO'] = 0
df['CL'] = 0
# Entities in jnlpba
df['DNA'] = 0
df['CELL_TYPE'] = 0
df['CELL_LINE'] = 0
df['RNA'] = 0
df['PROTEIN'] = 0
# Entities in bc5cdr
df['DISEASE'] = 0
df['CHEMICAL'] = 0
# Entities in bionlp13cg
df['AMINO_ACID'] = 0
df['ANATOMICAL_SYSTEM'] = 0
df['CANCER'] = 0
df['CELL'] = 0
df['CELLULAR_COMPONENT'] = 0
df['DEVELOPING_ANATOMICAL_STRUCTURE'] = 0
df['GENE_OR_GENE_PRODUCT'] = 0
df['IMMATERIAL_ANATOMICAL_ENTITY'] = 0
df['MULTI_TISSUE_STRUCTURE'] = 0
df['ORGAN'] = 0
df['ORGANISM'] = 0
df['ORGANISM_SUBDIVISION'] = 0
df['ORGANISM_SUBSTANCE'] = 0
df['PATHOLOGICAL_FORMATION'] = 0
df['SIMPLE_CHEMICAL'] = 0
df['TISSUE'] = 0
# Entities in custom medical-ner
df['Pathogen'] = 0
df['MedicalCondition'] = 0
df['Medicine'] = 0

def get_sci_entities(row):
    tokens1 = nlp1(''.join(str(row['text'])))
    tokens2 = nlp2(''.join(str(row['text'])))
    tokens3 = nlp3(''.join(str(row['text'])))
    tokens4 = nlp4(''.join(str(row['text'])))
    tokens5 = nlp5(''.join(str(row['text'])))
    for ent in tokens1.ents: # Entities in craft
        if ent.label_ == 'GGP':
            row['GGP'] += 1
        if ent.label_ == 'SO':
            row['SO'] += 1
        if ent.label_ == 'TAXON':
            row['TAXON'] += 1
        if ent.label_ == 'CHEBI':
            row['CHEBI'] += 1
        if ent.label_ == 'GO':
            row['GO'] += 1
        if ent.label_ == 'CL':
            row['CL'] += 1
    for ent in tokens2.ents: # Entities in jnlpba
        if ent.label_ == 'DNA':
            row['DNA'] += 1
        if ent.label_ == 'CELL_TYPE':
            row['CELL_TYPE'] += 1
        if ent.label_ == 'CELL_LINE':
            row['CELL_LINE'] += 1
        if ent.label_ == 'RNA':
            row['RNA'] += 1
        if ent.label_ == 'PROTEIN':
            row['PROTEIN'] += 1
    for ent in tokens3.ents: # Entities in bc5cdr
        if ent.label_ == 'DISEASE':
            row['DISEASE'] += 1
        if ent.label_ == 'CHEMICAL':
            row['CHEMICAL'] += 1
    for ent in tokens4.ents: # Entities in bionlp13cg
        if ent.label_ == 'AMINO_ACID':
            row['AMINO_ACID'] += 1
        if ent.label_ == 'ANATOMICAL_SYSTEM':
            row['ANATOMICAL_SYSTEM'] += 1
        if ent.label_ == 'CANCER':
            row['CANCER'] += 1
        if ent.label_ == 'CELL':
            row['CELL'] += 1
        if ent.label_ == 'CELLULAR_COMPONENT':
            row['CELLULAR_COMPONENT'] += 1
        if ent.label_ == 'DEVELOPING_ANATOMICAL_STRUCTURE':
            row['DEVELOPING_ANATOMICAL_STRUCTURE'] += 1
        if ent.label_ == 'GENE_OR_GENE_PRODUCT':
            row['GENE_OR_GENE_PRODUCT'] += 1
        if ent.label_ == 'IMMATERIAL_ANATOMICAL_ENTITY':
            row['IMMATERIAL_ANATOMICAL_ENTITY'] += 1
        if ent.label_ == 'MULTI_TISSUE_STRUCTURE':
            row['MULTI_TISSUE_STRUCTURE'] += 1
        if ent.label_ == 'ORGAN':
            row['ORGAN'] += 1
        if ent.label_ == 'ORGANISM':
            row['ORGANISM'] += 1
        if ent.label_ == 'ORGANISM_SUBDIVISION':
            row['ORGANISM_SUBDIVISION'] += 1
        if ent.label_ == 'ORGANISM_SUBSTANCE, ':
            row['ORGANISM_SUBSTANCE, '] += 1
        if ent.label_ == 'PATHOLOGICAL_FORMATION':
            row['PATHOLOGICAL_FORMATION'] += 1
        if ent.label_ == 'SIMPLE_CHEMICAL':
            row['SIMPLE_CHEMICAL'] += 1
        if ent.label_ == 'TISSUE':
            row['TISSUE'] += 1
    for ent in tokens5.ents: # Entities in custom medical-ner
        if ent.label_ == 'Pathogen':
            row['Pathogen'] += 1
        if ent.label_ == 'MedicalCondition':
            row['MedicalCondition'] += 1
        if ent.label_ == 'Medicine':
            row['Medicine'] += 1
            
    return row

df_features = df.apply(lambda row: get_sci_entities(row), axis=1)
```

```python
df_features = df_features.drop(columns=['title', 'text', 'source', 'label'])
```
