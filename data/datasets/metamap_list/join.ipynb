```python
import pandas as pd
from functools import reduce
```

```python
df1 = pd.read_csv('./metamap_features_1.csv')
df2 = pd.read_csv('./metamap_features_2.csv')
df3_1 = pd.read_csv('./metamap_features_3_1.csv')
df3_2 = pd.read_csv('./metamap_features_3_2.csv')
df3_3 = pd.read_csv('./metamap_features_3_3.csv')
df4 = pd.read_csv('./metamap_features_4.csv')
df5 = pd.read_csv('./metamap_features_5.csv')
df6 = pd.read_csv('./metamap_features_6.csv')
df7 = pd.read_csv('./metamap_features_7.csv')
df8 = pd.read_csv('./metamap_features_8.csv')
df9 = pd.read_csv('./metamap_features_9.csv')
df10 = pd.read_csv('./metamap_features_10.csv')
df11 = pd.read_csv('./metamap_features_11.csv')
df12 = pd.read_csv('./metamap_features_12.csv')
df13 = pd.read_csv('./metamap_features_13.csv')
```

```python
dataframes = [df1, df2, df3_1, df3_2, df3_3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13]
```

```python
def clean_df(df):
    df['index'] = df.index
    label = df[['index', 'label']]
    df = df.drop(columns=['label'])

    frames = [df, label]
    df_new = reduce(lambda  left,right: pd.merge(left,right,on=['index'],
                                        how='inner'), frames)
    return df_new
```

```python
df1 = clean_df(df1)
df2 = clean_df(df2)
df3_1 = clean_df(df3_1)
df3_2 = clean_df(df3_2)
df3_3 = clean_df(df3_3)
df4 = clean_df(df4)
df5 = clean_df(df5)
df6 = clean_df(df6)
df7 = clean_df(df7)
df8 = clean_df(df8)
df9 = clean_df(df9)
df10 = clean_df(df10)
df11 = clean_df(df11)
df12 = clean_df(df12)
df13 = clean_df(df13)
```

```python
df = pd.concat(dataframes)
df.reset_index

# print(len(df))
df.head()
```

```python
# Fix label problem
df['label'] = df['label'].str.lower()
df.head()
```

```python
# Encode categorical variables
df = pd.get_dummies(df, columns=['label'])
df.head()
```

```python
# Drop unnecessary columns
df = df.drop(columns=['label_true', 'index'])
df.head()
```

```python
df.to_csv('../10.csv', index=False)
```
