```python
import pandas as pd
from functools import reduce
```

```python
# Read in dataframes to concatenate
df1 = pd.read_csv('../data/datasets/4.csv')
df2 = pd.read_csv('../data/datasets/8.csv')

df1['index'] = df1.index
df2['index'] = df2.index
```

```python
df1 = df1.drop(columns=['label_fake'])
```

```python
dataframes = [df1, df2]

df6 = reduce(lambda  left,right: pd.merge(left,right,on=['index'],
                                            how='outer'), dataframes)

df6 = df6.drop(columns=['index'])
df6.head()
```

```python
# df10 = df_features_3[['Pathogen', 'MedicalCondition', 'Medicine', 'label_fake']]
df6.to_csv('../data/datasets/6.csv', index=False)
```
